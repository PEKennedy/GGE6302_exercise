```python
import os
import csv
import argparse
import matplotlib.pyplot as plt
import math
import numpy as np
%matplotlib inline

Sidney = "TideData\wlo_Sidney.tid"
Victoria = "TideData\wlo_Victoria Harbour.tid"
```

```python
def analyze_tide(file_name):
    with open(file_name) as file:
        min_level = 99999999
        max_level = -99999999
        ave_level = 0
        levels = []

        # .tid files modified to be csv
        for row in csv.DictReader(file, delimiter=" "):
            level = float(row["water_level"])
            
            min_level = min(level, min_level)
            max_level = max(level, max_level)
            levels.append(level)
        ave_level = np.mean(levels)
        print(min_level)
        print(max_level)
        #print(levels)
        print(ave_level)

        
```

```python
analyze_tide(Sidney)
```

```python
analyze_tide(Victoria)
```
